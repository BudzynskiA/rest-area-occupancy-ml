```python
# Required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

# Load and clean the dataset
df = pd.read_csv("rest_area_occupancy_merged.csv")
df_clean = df.dropna()

# Encode categorical features
df_encoded = df_clean.copy()
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    if col != "Occupancy":
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
        label_encoders[col] = le

# Encode the target variable
le_target = LabelEncoder()
y = le_target.fit_transform(df_encoded["Occupancy"])
X = df_encoded.drop(columns=["Occupancy"])

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define models
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=10000, random_state=42),
    "XGBoost": XGBClassifier(eval_metric="mlogloss", random_state=42)
}

# Evaluate models using weighted F1-score
f1_results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, target_names=le_target.classes_, output_dict=True)
    f1_results[name] = report["weighted avg"]["f1-score"]

# Convert results to DataFrame and sort
f1_df = pd.DataFrame.from_dict(f1_results, orient="index", columns=["F1-score"]).sort_values(by="F1-score", ascending=False)

# Plot with value labels
plt.figure(figsize=(8, 5))
bars = sns.barplot(x=f1_df["F1-score"].round(3), y=f1_df.index, color="steelblue")
plt.title("Weighted F1-score comparison across models")
plt.xlabel("F1-score")
plt.ylabel("Model")
plt.xlim(0, 1)
plt.grid(axis='x', linestyle='--', linewidth=0.5)

# Add value labels to bars
for bar in bars.containers[0]:
    width = bar.get_width()
    plt.text(width + 0.01, bar.get_y() + bar.get_height() / 2,
             f"{width:.2f}", va='center')

plt.tight_layout()
plt.show()

# Print F1-scores table
print(f1_df.round(3))
```

```python
from sklearn.metrics import classification_report

# Redefine models with balanced class weight for Logistic Regression
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=10000, class_weight='balanced', random_state=42),
    "XGBoost": XGBClassifier(eval_metric="mlogloss", random_state=42)
}

# Collect F1-scores per class for each model
f1_class_results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, target_names=le_target.classes_, output_dict=True)
    f1_class_results[name] = {
        cls: round(metrics["f1-score"], 3) for cls, metrics in report.items() if cls in le_target.classes_
    }

# Convert results to DataFrame
f1_class_df = pd.DataFrame(f1_class_results).T
f1_class_df = f1_class_df[le_target.classes_]  # consistent column order

# Display table
print("F1-score per class for each model:")
print(f1_class_df)

# Horizontal grouped bar chart
f1_class_df.plot(kind='barh', figsize=(10, 6))
plt.title("F1-score per occupancy class by model")
plt.xlabel("F1-score")
plt.ylabel("Model")
plt.xlim(0, 1)
plt.grid(axis='x', linestyle='--', linewidth=0.5)

# Improved legend position
plt.legend(title="Class", loc='upper right', bbox_to_anchor=(1.15, 1))

plt.tight_layout()
plt.show()
```
