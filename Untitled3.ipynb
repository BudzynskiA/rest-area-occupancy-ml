```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming it's in the same directory)
df = pd.read_csv("rest_area_occupancy_merged.csv")

# List of facility columns to analyze
facility_columns = [
    "Restauracja /Bistro", "Miejsca noclegowe", "Toalety", "Prysznic", "Stacja paliw",
    "Ładowanie pojazdów elektrycznych", "Myjnia", "Warsztat", "Ochrona", 
    "Monitoring wideo", "Ogrodzenie", "Oświetlenie"
]

# Rename columns to English for better readability
rename_dict = {
    "Restauracja /Bistro": "Restaurant / Bistro",
    "Miejsca noclegowe": "Accommodation",
    "Toalety": "Toilets",
    "Prysznic": "Shower",
    "Stacja paliw": "Fuel Station",
    "Ładowanie pojazdów elektrycznych": "EV Charging",
    "Myjnia": "Car Wash",
    "Warsztat": "Workshop",
    "Ochrona": "Security",
    "Monitoring wideo": "Video Monitoring",
    "Ogrodzenie": "Fencing",
    "Oświetlenie": "Lighting"
}
df.rename(columns=rename_dict, inplace=True)
facility_columns = [rename_dict.get(col, col) for col in facility_columns]

# Prepare data for plotting
occupancy_levels = ["low", "medium", "high"]
plot_data = []

for feature in facility_columns:
    subset = df[df[feature] == "tak"]
    counts = subset["Occupancy"].value_counts(normalize=True).reindex(occupancy_levels, fill_value=0) * 100
    plot_data.append(counts)

plot_df = pd.DataFrame(plot_data, index=facility_columns)

# Plot: Horizontal grouped bar chart
fig, ax = plt.subplots(figsize=(10, 7))
bar_width = 0.25
positions = range(len(plot_df))

for i, occ in enumerate(occupancy_levels):
    ax.barh(
        [p + bar_width * i for p in positions],
        plot_df[occ],
        height=bar_width,
        label=occ.capitalize(),
        color={"low": "green", "medium": "yellow", "high": "red"}[occ]
    )

# Formatting
ax.set_yticks([p + bar_width for p in positions])
ax.set_yticklabels(plot_df.index)
ax.set_xlabel("Percentage of Observations")
#ax.set_title("Occupancy Distribution by Facility (Presence = Yes)")
ax.legend(title="Occupancy Level")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
