```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("rest_area_occupancy_merged.csv")
df = df.dropna()

# Translate column names
translation_dict = {
    'Myjnia': 'Car Wash',
    'Ochrona': 'Security',
    'Stacja paliw': 'Fuel Station',
    'Ładowanie pojazdów elektrycznych': 'EV Charging',
    'Miejsca noclegowe': 'Accommodation',
    'Prysznic': 'Shower',
    'Monitoring wideo': 'Video Monitoring',
    'Możliwość tankowania wodorem': 'Hydrogen Refueling',
    'Możliwość tankowania LNG': 'LNG Refueling',
    'Możliwość tankowania CNG': 'CNG Refueling',
    'Oświetlenie': 'Lighting',
    'Ogrodzenie': 'Fencing',
    'Toalety': 'Toilets',
    'Restauracja /bistro': 'Restaurant / Bistro',
    'Warsztat': 'Workshop'
}
df.rename(columns=translation_dict, inplace=True)

# Map occupancy to numeric scale
occupancy_map = {'low': 1, 'medium': 2, 'high': 3}
df['Occupancy_num'] = df['Occupancy'].map(occupancy_map)

# Select relevant facility features
features = list(translation_dict.values())

# Compute average occupancy for presence/absence of each feature
with_feature = []
without_feature = []
valid_features = []

for col in features:
    if col in df.columns:
        mean_yes = df[df[col] == "tak"]['Occupancy_num'].mean()
        mean_no = df[df[col] == "nie"]['Occupancy_num'].mean()
        if not np.isnan(mean_yes) and not np.isnan(mean_no):
            with_feature.append(mean_yes)
            without_feature.append(mean_no)
            valid_features.append(col)

# Plotting setup
x = np.arange(len(valid_features))
width = 0.4

plt.figure(figsize=(10, 7))
bar1 = plt.barh(x - width/2, with_feature, height=width, label="Yes", color="green")
bar2 = plt.barh(x + width/2, without_feature, height=width, label="No", color="lightgray")

# Add numerical labels
for bars in [bar1, bar2]:
    for bar in bars:
        plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
                 f"{bar.get_width():.2f}", va='center', fontsize=8)

# Customize plot
plt.yticks(x, valid_features)
plt.xlabel("Average occupancy level (1=low, 2=medium, 3=high)")
plt.title("Average occupancy level for rest areas with and without each feature")
plt.legend(title="Presence of feature")
plt.xlim(1, 3)
plt.tight_layout()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()

```
