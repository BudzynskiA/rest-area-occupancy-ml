```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the processed dataset
df = pd.read_csv("rest_area_occupancy_merged.csv")
```

```python
# Normalize the Hour column to HH:MM format (force HH:MM only)
df["Hour"] = df["Hour"].astype(str).str.extract(r"(\d{2}:\d{2})")[0]
```

```python
# --- Cell 1: Dataset info ---
print("Basic info about the loaded DataFrame:")
print(df.info())
print("\nHead of the DataFrame:")
print(df.head())
print("\nDescriptive statistics:")
print(df.describe(include='all'))
```

```python
# --- Cell 2: Overall occupancy distribution ---
occupancy_counts = df["Occupancy"].value_counts().reindex(["low", "medium", "high"])
bars = occupancy_counts.plot(kind="bar", color=["#66c2a5", "#fc8d62", "#8da0cb"])
#plt.title("Overall Occupancy Distribution")
plt.xlabel("Occupancy Level")
plt.ylabel("Number of Observations")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels with proper spacing and limit y-axis for margin
for i, v in enumerate(occupancy_counts):
    plt.text(i, v * 1.01, str(v), ha='center', va='bottom')

# Add extra margin to top of y-axis
plt.ylim(top=max(occupancy_counts) * 1.10)
plt.tight_layout()
plt.show()
```

```python
# --- Cell 3: Occupancy by hour (line plot) ---
hourly_occupancy = df.groupby(["Hour", "Occupancy"]).size().unstack().reindex(columns=["low", "medium", "high"])
hourly_occupancy = hourly_occupancy.fillna(0).sort_index()

hourly_occupancy.plot(kind="line", marker="o", figsize=(12, 5), color=["#66c2a5", "#fc8d62", "#8da0cb"])
#plt.title("Hourly Occupancy Trends")
plt.xlabel("Hour")
plt.ylabel("Number of Observations")
plt.xticks(range(0, 24))
plt.legend(title="Occupancy", loc="upper right")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
